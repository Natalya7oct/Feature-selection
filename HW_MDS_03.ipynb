Придумайте функции со следующими свойствами (в a) b) c) область определения X любая, какая вам удобна, главное - функция с нужными свойствами):

d*) область определения функции - 1 точка

```
import numpy as np
import math
import matplotlib.pyplot as plt
import pylab
```

```
x=np.linspace(-50,50,10000)
```

```
def f_a(x):
  return math.sin(x)
```

```
y = []
for i in x:
  y.append(f_a(i))
pylab.plot(x,y)
```

```
def f_b(x):
  return (x**3)/3-100*x
y = []
for i in x:
  y.append(f_b(i))
pylab.plot(x,y)
```

```
x=np.linspace(-15,15,100)
```

```
def f_c(x):
  return (x**4)/4-10*(x**3)/3-100*(x**2)/2+1000*x
y = []
for i in x:
  y.append(f_c(i))
pylab.plot(x,y)
```

```
def f_d(x):
  return math.sqrt(x)-math.sqrt(-x)
```

```
x = 0
y = f_d(x)
pylab.plot(x,y)
pylab.plot(x, f_d(x), 'rd')
```

Задание 3:
Используя numpy, посчитайте значение 3-ей и 4-ой производных функций в точке:

$f(x) = x^5+4sin(2x)+cos(3x+3)$ в точке $x=1$

```
def f_3(x):
  return x**5+4*math.sin(2*x)+math.cos(3*x+3)
```

```
from scipy.misc import derivative
print(derivative(f_3, 1.0, 3))
print(derivative(f_3, 1.0, 4))
```
