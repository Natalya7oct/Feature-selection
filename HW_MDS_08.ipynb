```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts

%matplotlib inline
```

Задание 1
Получить интуицию, на что влияют параметры распределений. Построить по 5 функций плотностей с разными параметрами для каждого распределения: нормальное, экспоненциальное, Стьюдента на одном графике.

```
fig, (ax1, ax2, ax3) = plt.subplots(
    nrows=1, ncols=3,
    figsize=(30, 5)
)

x = np.linspace(-10,10,100)
size=10

mu = -1
sigma = 1
norm_rv = sts.norm(loc=mu, scale=sigma)
norm_rv.rvs(size=size)
pdf = norm_rv.pdf(x)
ax1.plot(x, pdf, label='mu=-1;sigma=1')

mu = 1
sigma = 1
norm_rv = sts.norm(loc=mu, scale=sigma)
norm_rv.rvs(size=size)
pdf = norm_rv.pdf(x)
ax1.plot(x, pdf, label='mu=1;sigma=1')

mu = -1
sigma = 2
norm_rv = sts.norm(loc=mu, scale=sigma)
norm_rv.rvs(size=size)
pdf = norm_rv.pdf(x)
ax1.plot(x, pdf, label='mu=-1;sigma=2')

mu = 0
sigma = 3
norm_rv = sts.norm(loc=mu, scale=sigma)
norm_rv.rvs(size=size)
pdf = norm_rv.pdf(x)
ax1.plot(x, pdf, label='mu=0;sigma=3')

mu = 5
sigma = 2
norm_rv = sts.norm(loc=mu, scale=sigma)
norm_rv.rvs(size=size)
pdf = norm_rv.pdf(x)
ax1.plot(x, pdf, label='mu=5;sigma=2')

x = np.linspace(-10,10,100)

lam = 0.01
expon_rv = sts.expon(lam)
expon_rv.rvs(size=size)
pdf = expon_rv.pdf(x)
ax2.plot(x, pdf, label='lam = 0.01')


lam = 0.1
expon_rv = sts.expon(lam)
expon_rv.rvs(size=size)
pdf = expon_rv.pdf(x)
ax2.plot(x, pdf, label='lam = 0.1')

lam = 1
expon_rv = sts.expon(lam)
expon_rv.rvs(size=size)
pdf = expon_rv.pdf(x)
ax2.plot(x, pdf, label='lam = 1')

lam = 3
expon_rv = sts.expon(lam)
expon_rv.rvs(size=size)
pdf = expon_rv.pdf(x)
ax2.plot(x, pdf, label='lam = 3')

lam = 5
expon_rv = sts.expon(lam)
expon_rv.rvs(size=size)
pdf = expon_rv.pdf(x)
ax2.plot(x, pdf, label='lam = 5')


n = 0.01
t_rv = sts.t(n)
t_rv.rvs(size=size)
pdf = t_rv.pdf(x)
ax3.plot(x, pdf, label='n = 0.01')

n = 0.1
t_rv = sts.t(n)
t_rv.rvs(size=size)
pdf = t_rv.pdf(x)
ax3.plot(x, pdf, label='n = 0.1')

n = 1
t_rv = sts.t(n)
t_rv.rvs(size=size)
pdf = t_rv.pdf(x)
ax3.plot(x, pdf, label='n = 1')

n = 10
t_rv = sts.t(n)
t_rv.rvs(size=size)
pdf = t_rv.pdf(x)
ax3.plot(x, pdf, label='n = 10')

n = 100
t_rv = sts.t(n)
t_rv.rvs(size=size)
pdf = t_rv.pdf(x)
ax3.plot(x, pdf, label='n = 100')

ax1.set_ylabel('$f(x)$')
ax1.set_xlabel('$x$')
ax1.set_title('Нормальное распределение')
ax1.legend(loc='upper right')

ax2.set_ylabel('$f(x)$')
ax2.set_xlabel('$x$')
ax2.set_title('Экспоненциальное распределение')
ax2.legend(loc='upper right')

ax3.set_ylabel('$f(x)$')
ax3.set_xlabel('$x$')
ax3.set_title('Распределение Стъюдента')
ax3.legend(loc='upper right')

fig.show()
```
